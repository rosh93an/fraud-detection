```python
fraud_df=df[df['Class']==1]
normal_df=df[df['Class']==0]

#***---- Fraudtransactions******
print(fraud_df['Amount'].describe())

#*****-----Normaltransactions---****
print(normal_df['Amount'].describe())
```

```python
print(f"\nAverage fraud amount:  €{fraud_df['Amount'].mean():.2f}")
print(f"Average normal amount: €{normal_df['Amount'].mean():.2f}")
print(f"Max fraud amount:      €{fraud_df['Amount'].max():.2f}")
print(f"Max normal amount:     €{normal_df['Amount'].max():.2f}")
```

FRAUD transactions:
Mean: €122 — fraudsters prefer medium amounts
75th percentile: €105 — 75% of fraud is UNDER €105
Max: €2125 — fraud rarely goes very high

NORMAL transactions:
Mean: €88 — actually LOWER than fraud mean
Max: €25,691 — legitimate transactions go much higher
75th percentile: €77

Fraud sweet spot = €1 to €105
Why? Fraudsters test with small amounts first
     Then do medium transactions to avoid detection
     They AVOID large amounts — triggers bank alerts
     This is called "structuring" in financial crime

Normal max = €25,691
Normal 75th percentile = €77
Normal mean = €88

So someone spent €25,691 when 75% of people
spend under €77 — YES that is an outlier

But here's the important question:
Is that outlier fraud or legitimate?

Answer: Could be either
→ Rich person buying jewellery = legitimate outlier
→ Stolen card used for big purchase = fraud outlier

This is why Amount alone cannot detect fraud
We need V1-V28 features combined with Amount

look upper both charts closely Question: Normal transactions show two peaks in the time chart. Why do you think that happens?
Time is in seconds from start of data collection
172,800 seconds = 48 hours (2 days of data)

Normal transactions two peaks:
Peak 1: ~75,000 seconds = ~20 hours (Day 1 evening)
Dip:    ~100,000-125,000 = night time (people sleeping)
Peak 2: ~150,000 seconds = next day business hours

This is DAILY PATTERN of human behavior:
Morning → transactions rise
Evening → peak activity  
Night   → dip (people sleeping)
Next day → rises again

Fraud chart is RANDOM — no clear pattern
Fraudsters don't follow business hours
They operate 24/7 across both days
This randomness IS the pattern — 
fraud has no day/night preference

```python
# Which features correlate most with fraud?
plt.figure(figsize=(8, 6))

# Correlation of all features with Class
correlations = df.corr()['Class'].drop('Class').sort_values()

# Plot top 10 positive and negative correlations
top_corr = pd.concat([correlations.head(10), 
                       correlations.tail(10)])

colors = ['#e74c3c' if x > 0 else '#2ecc71' 
          for x in top_corr.values]

top_corr.plot(kind='barh', 
              color=colors,
              figsize=(10, 6),
              edgecolor='black')

plt.title('Top Feature Correlations with Fraud', 
          fontweight='bold', fontsize=14)
plt.xlabel('Correlation Coefficient')
plt.axvline(x=0, color='black', linewidth=0.5)
plt.tight_layout()
plt.savefig('notebooks/correlations.png')
plt.show()

print("\nTop 5 features most correlated with FRAUD:")
print(correlations.tail(5))
print("\nTop 5 features most negatively correlated with FRAUD:")
print(correlations.head(5))
```

Why normal max (€25,691) not visible clearly?
Because 200,000+ transactions are between €0-500
That giant bar at left makes everything else 
look flat — this is called a RIGHT SKEWED distribution

Both fraud and normal are right skewed:
→ Most transactions are small amounts
→ Very few large transactions exist
→ Large amounts are outliers pulling the mean right

Fraud mean (€122) appears AFTER the biggest bar
→ Most fraud is small BUT some large frauds 
  pull the average up to €122

Normal mean (€88) is AT the biggest bar
→ Most normal transactions cluster around €88

RED bars (positive correlation = more = more fraud):

GREEN bars (negative correlation = less = more fraud):

```python
IMPORTANT_FEATURES =['V11', 'V4', 'V2', 'V17', 'V14', 'Amount']
print("===fraud cases-important features")
print(fraud_df[IMPORTANT_FEATURES].describe().round(2))

print("====normalcases-important features")
print(normal_df[IMPORTANT_FEATURES].describe().round(2))
```

Normal transactions:
V11 mean = -0.01  ← almost exactly 0
V11 std  =  1.00  ← almost exactly 1

This is textbook PCA normalization
Mean ≈ 0, Std ≈ 1 for the FULL dataset

PCA was applied on the FULL dataset
(284,807 transactions combined)

Full dataset mean ≈ 0  ✅
Full dataset std  ≈ 1  ✅

BUT when you SEPARATE fraud from normal:
Normal (99.83%) pulls the combined mean to ≈ 0
Fraud  (0.17%)  gets "left out" — its natural 
                pattern is revealed

Fraud V17 mean = -6.67
This means fraud transactions have 
CONSISTENTLY LOW V17 values
Normal V17 mean = 0.01

The separation between -6.67 and 0.01 
IS the fraud signal hiding in the data

V17 fraud:
  25% = -11.95  ← very negative
  50% = -5.30   ← very negative  
  75% = -1.34   ← still negative

V17 normal:
  25% = -0.48   ← near zero
  50% = -0.06   ← near zero
  75% =  0.40   ← near zero

The model learns this rule:
"If V17 < -3 AND V14 < -5 → very likely fraud"

These features create a BOUNDARY that 
separates fraud from normal
That boundary is what the model learns
